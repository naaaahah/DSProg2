```python
#スクレイピング
#ライブラリのインポート
import requests
from bs4 import BeautifulSoup
```

```python
reget = requests.get("https://www.mcsg.co.jp/kentatsu/health-care/36561")
soup = BeautifulSoup(reget.text, 'html.parser')
print(soup)
```

```python
#n代男性のデータを格納
list_20s_man = []
list_30s_man = []
list_40s_man = []
list_50s_man = []
list_60s_man = []
list_70s_man = []
all_list = [list_20s_man, list_30s_man, list_40s_man, list_50s_man, list_60s_man, list_70s_man]

nth = 2
for list in all_list:
    for i in range(3,9):
        data = soup.select_one(f"#top > div.wrapper-l-wrapper > div > main > div:nth-child(1) > div.postContents > section > div:nth-child(85) > div > table > tbody > tr:nth-child({i}) > td:nth-child({nth}) > span")
        data = data.text
        list.append(data)
    nth += 1
print(all_list)
```
